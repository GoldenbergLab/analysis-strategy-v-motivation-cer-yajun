```python
from transformers import BertTokenizer
from transformers import BertModel
from tqdm import tqdm


import torch
import pandas as pd
import csv
import numpy
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
```

```python
text_dat = pd.read_csv("text_data.csv")
text_dat.head()
text_dat.shape
```

```python
#embeds = list()
i = 0
for text in tqdm(text_dat["clean_text"]):
    try:
        input_ids = torch.tensor(tokenizer.encode(text)).unsqueeze(0)  # Batch size 1
        outputs = model(input_ids)[1].detach().numpy()[0] # pooler output
        outputs = numpy.append(outputs, i)
        i += 1
        with open('embeds.csv','a') as f:
            write = csv.writer(f)
            write.writerow(outputs)
    except:
        pass

```

```python
embed_dat = pd.read_csv("embeds.csv", header = None)

embed_dat.shape

```

```python
embed_dat.head()
```

```python
num_embeds = embed_dat.shape[1]
embed_names = list()
for i in tqdm(range(num_embeds-1)):
    embed_names.append("embed" + str(i))
    
embed_names.append("num")
    
embed_dat.columns = embed_names
embed_dat.head()
```

```python
text_dat = text_dat.assign(num = range(text_dat.shape[0]))

text_embed = pd.merge(text_dat, embed_dat, on = "num")
text_embed.to_csv("embed_dat.csv")
```
